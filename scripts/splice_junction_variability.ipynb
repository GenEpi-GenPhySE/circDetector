```python
# Imports:
import seaborn as sns
import numpy as np
import os
from dataclasses import dataclass
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from typing import *
import warnings
warnings.filterwarnings("ignore")
from IPython.display import display, HTML
    
# Functions:
def read_file(file):
    df = pd.read_table(file, sep = '\t')
    df = df.replace(np.nan, 0, regex=True) 
    df["start"] = df["start"].astype(str).astype(int)
    df["end"] = df["end"].astype(str).astype(int)
    #df.astype({"start": int, "end": int, "start_i": int, "end_i": int})
    return df

def compute_variability(df):
    nb_ccr_tot = df["nb_ccr"].sum()
    df["diff_start"] = abs(df["start_i"] - df["start"])
    df["diff_end"] = abs(df["end_i"] - df["end"])
    sum_diff_starts = df["diff_start"].sum()
    sum_diff_ends = df["diff_end"].sum()
    var_start = round(sum_diff_starts/nb_ccr_tot, 4)
    var_end = round(sum_diff_ends/nb_ccr_tot, 4)
    return "\t".join(map(str, [var_start, var_end]))
```

```python
# Print dataframe:
df = read_file("../results/cow-liver-B004/annotation_circRNAs.out")
display(df)
```

```python
# Compute the splice junctions variability:
var = compute_variability(df)
print(var)
```
